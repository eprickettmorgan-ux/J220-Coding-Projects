### Show me the MEDIAN enrollment of WOMEN by STATE sorted in order from highest to lowest.
Note, I don't need the names of individual colleges anymore, just the states.

```python
MedianWomen = df.groupby('state')['women'].median()
MedianWomen = MedianWomen.sort_values(ascending=False)
print(MedianWomen)
```

### Same chart as above, except clip to the top 10 and make a horizontal bar chart

Don't worry about sorting from highest to lowest any more.

```python
MedianWomen = df.groupby('state')['women'].median()
MedianWomenChart = MedianWomen.sort_values(ascending=False).head(10)
MedianWomenChart.plot.bar(x='state', y='women', rot=90)
plt.show()
```

### Show me a list of colleges where women are the highest percentage

`percent_women = women / enrollment`

```python
pctwmndf = df.assign(percent_women = (df['women']/df['enrollment'])*100)
pctwmndf.sort_values(by='percent_women', ascending=False)
```

### Show me a list of colleges where international (nonresident) students are the highest percentage of enrollment

`percent_international = nonresident / enrollment`

```python
pctintdf = df.assign(percent_international = (df['nonresident']/df['enrollment'])*100)
pctintdf.sort_values(by='percent_international', ascending=False)
```

### List the top 10 states that have the highest AVERAGE (mean) of enrolled students?

```python
MeanEnrollment = df.groupby('state')['enrollment'].mean()
StateMeanEnrollment = MeanEnrollment.sort_values(ascending=False).head(10)
print(StateMeanEnrollment)
```

### Create several charts showing some interesting elements you find in these data

Some ideas might be comparing UC Berkeley to other UCs, looking at Ivy League comparisons, running some regressions

```python
StateEnrollment = df.groupby('state')['enrollment'].sum()
Top10StateEnrollment = StateEnrollment.sort_values(ascending=False).head(10)
Top10StateEnrollment.plot.bar(x='state', y='enrollment', rot=90)
plt.show()
print(Top10StateEnrollment)
```

```python
BlackEnrollment = df.groupby('state')['black'].sum()
Top10BlackEnrollment = BlackEnrollment.sort_values(ascending=False).head(10)
Top10BlackEnrollment.plot.bar(x='state', y='black', rot=90)
plt.show()
print(Top10BlackEnrollment)
```

```python
NatEnrollment = df.groupby('state')['american_indian_or_alaska_native'].sum()
Top10NatEnrollment = NatEnrollment.sort_values(ascending=False).head(10)
Top10NatEnrollment.plot.bar(x='state', y='american_indian_or_alaska_native', rot=90)
plt.show()
print(Top10NatEnrollment)
```
